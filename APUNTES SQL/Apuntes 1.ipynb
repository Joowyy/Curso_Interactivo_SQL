<p class=MsoNormal align=center style='text-align:center;border:none;
padding:0cm'><b><span style='font-size:16.0pt;line-height:107%;color:#0070C0'>Sistema
Gestor de Base de datos o SGBD</span></b></p>

</div>

<p class=MsoNormal style='text-align:justify'>&nbsp;</p>

<p class=MsoNormal style='text-align:justify'>Un Sistema de Gestión de Bases de
Datos es un tipo de software muy específico dedicado a servir de interfaz entre
la base de datos, el usuario y las aplicaciones que la utilizan; o lo que es lo
mismo, una agrupación de programas que sirven para definir, construir y
manipular una base de datos, permitiendo así almacenar y posteriormente acceder
a los datos de forma rápida y estructurada, garantizando además la seguridad de
los mismos.</p>

<p class=MsoNormal align=center style='text-align:center'><span lang=ES-TRAD><a
href="./imagenes/sgbd.png"
target="_blank"><span lang=ES style='color:windowtext;text-decoration:none'><img
border=0 width=486 height=301 id="Imagen 8"
src=./imagenes/image001.png"></span></a></span></p>

<p class=MsoNormal><i>Sistema Gestor Base de Datos</i></p>

<p class=MsoNormal>El éxito del SGBD reside en mantener la seguridad e
integridad de los datos. Lógicamente tiene que proporcionar herramientas a los
distintos usuarios. Entre las herramientas que proporciona están:</p>

<ul style='margin-top:0cm' type=disc>
 <li class=MsoNormal>Herramientas para la&nbsp;<b>creación y especificación de
     los datos</b>. Así como la estructura de la base de datos.</li>
 <li class=MsoNormal>Herramientas para&nbsp;<b>administrar y crear la
     estructura física</b>&nbsp;requerida en las unidades de almacenamiento.</li>
 <li class=MsoNormal>Herramientas para la&nbsp;<b>manipulación de los datos</b>&nbsp;de
     las bases de datos, para añadir, modificar, suprimir o consultar datos.</li>
 <li class=MsoNormal>Herramientas para la&nbsp;<b>creación de copias de
     seguridad</b>.</li>
 <li class=MsoNormal>Herramientas para la&nbsp;<b>gestión de la comunicación</b>&nbsp;de
     la base de datos.</li>
 <li class=MsoNormal>Herramientas para la&nbsp;<b>creación de aplicaciones</b>&nbsp;que
     utilicen esquemas externos de los datos.</li>
 <li class=MsoNormal>Herramientas de&nbsp;<b>instalación de la base de datos</b>.</li>
 <li class=MsoNormal>Herramientas para la&nbsp;<b>exportación e importación de
     datos</b>.</li>
</ul>

<p class=MsoNormal><a name="caracter%C3%ADsticasdeseablesenunsgbd"></a><a
name="sistemagestordebasededatososgbd/caracter"></a><a name=toc3.1></a><b>Características
deseables en un SGBD</b></p>

<ul style='margin-top:0cm' type=disc>
 <li class=MsoNormal><b>Control de redundancia</b>. El SGBD consta de
     herramienta para crear restricciones para evitar duplicaciones.</li>
 <li class=MsoNormal style='text-align:justify'><b>Restricción de accesos no
     autorizados</b>. Muchos usuarios comparten la base pero no todos deben
     tener acceso a toda la información que ella contiene. El SGBD debe contar
     con un subsistema de seguridad y autorización que permita administrar
     restricciones y autorizaciones.</li>
 <li class=MsoNormal><b>Cumplimiento de las restricciones de integridad</b>.
     Restricciones de los datos que aplicaremos mediante el SGBD.</li>
 <li class=MsoNormal><b>Respaldo y recuperación</b>. Respaldo y recuperación se
     refiere a la recuperación ante un fallo de hardware o software del SGBD.
     Si por algún motivo el SGBD falla debe volver a la instancia anterior al
     fallo.</li>
</ul>

<div class=WordSection1>

<div style='border:solid #0070C0 2.25pt;padding:1.0pt 4.0pt 1.0pt 4.0pt'>

<p class=MsoNormal align=center style='text-align:center;border:none;
padding:0cm'><b><span style='font-size:16.0pt;line-height:107%;color:#0070C0'>Modelado
de bases de datos</span></b></p>

</div>
<br>
<p class=MsoNormal style='text-align:justify'>Las bases de datos aparecieron a
finales de la década de 1960, cuando surgió la necesidad de contar con un
sistema de administración de información flexible. Existen diferentes modelos
de DBMS, que se distinguen según cómo representan los datos almacenados.</p>

<p class=MsoNormal style='text-align:justify'>Un&nbsp;<b>modelo de datos</b>&nbsp;se
define como la estructura lógica que tiene la base de datos, es decir, como se
organizan los datos y las relaciones que existen entre ellos.</p>

<p class=MsoNormal><b>Tipos de modelos de bases de datos</b></p>

<ul style='margin-top:0cm' type=disc>
 <li class=MsoNormal style='text-align:justify'><b>Modelo Jerárquico</b>. Los
     datos se organizan jerárquicamente mediante un árbol invertido, donde cada
     registro tiene un nodo raiz del que surgen otros nodos. Este modelo
     utiliza punteros para navegar por los datos almacenados. Fue el primer
     modelo DBMS.</li>
</ul>

<p class=MsoNormal align=center style='text-align:center'><span lang=ES-TRAD><a
href="./imagenes/mod_jerar.png"
target="_blank"><span lang=ES style='color:windowtext;text-decoration:none'><img
border=0 width=325 height=155 id="Imagen 7"
src=./imagenes/image002.png"></span></a></span></p>

<p class=MsoNormal align=center style='text-align:center'><i>Modelo Jerárquico</i></p>

<p class=MsoNormal style='text-align:justify'>En la estructura de árbol (visto
al revés), un nodo padre de información puede tener varios hijos.</p>

<p class=MsoNormal style='text-align:justify'>El nodo que no tiene padres es
llamado&nbsp;<b>raíz</b>, y a los nodos que no tienen hijos se los conoce
como&nbsp;<b>hojas</b>.</p>

<p class=MsoNormal style='text-align:justify'>Las bases de datos jerárquicas
son especialmente útiles en el caso de aplicaciones que manejan un gran volumen
de información y datos muy compartidos permitiendo crear estructuras estables y
de gran rendimiento.</p>

<p class=MsoNormal style='text-align:justify'>Una de las principales
limitaciones de este modelo es su incapacidad de representar eficientemente la
redundancia de datos.</p>

<ul style='margin-top:0cm' type=disc>
 <li class=MsoNormal><b>Modelo de Red</b>. Parte del modelo jerárquico, y al
     igual que él, este modelo utiliza punteros hacia los datos almacenados.
     Sin embargo, no necesariamente utiliza una estructura de árbol invertido.</li>
</ul>

<p class=MsoNormal align=center style='text-align:center'><span lang=ES-TRAD><a
href="./imagenes/mod_red.png"
target="_blank"><span lang=ES style='color:windowtext;text-decoration:none'><img
border=0 width=567 height=275 id="Imagen 6"
src="./imagenes/image003.png"></span></a></span></p>

<p class=MsoNormal align=center style='text-align:center'><i>Modelo de Red</i></p>

<p class=MsoNormal style='text-align:justify'>Modifica el concepto de nodo: se
permite que un mismo nodo tenga varios padres (posibilidad no permitida en el
modelo jerárquico).</p>

<p class=MsoNormal style='text-align:justify'>Fue una gran mejora con respecto
al modelo jerárquico, ya que ofrecía una solución eficiente al problema de
redundancia de datos; pero, aun así, sigue siendo difícil administrar la
información.</p>

<ul style='margin-top:0cm' type=disc>
 <li class=MsoNormal><b>Modelo Relacional</b>. (RDBMS, Relational Database
     Management System [Sistema de administración de bases de datos
     relacionales]). Ordena los datos en tablas de dos dimensiones, compuestas
     de filas y columnas.</li>
</ul>

<p class=MsoNormal>Su idea fundamental es el uso de relaciones. Es el modelo
más utilizado en la actualidad para modelar problemas reales y administrar
datos dinámicamente. Los datos se manipulan según la teoría relacional de
matemáticas.</p>

<p class=MsoNormal align=center style='text-align:center'><span lang=ES-TRAD><a
href="./imagenes/mod_rel.png"
target="_blank"><span lang=ES style='color:windowtext;text-decoration:none'><img
border=0 width=567 height=435 id="Imagen 5"
src="./imagenes/image004.jpg"></span></a></span></p>

<p class=MsoNormal align=center style='text-align:center'><i>Modelo Relacional</i></p>

<p class=MsoNormal style='text-align:justify'>Creado en 1970 por Edgar Codd, no
tardó en consolidarse como un nuevo paradigma en los modelos de base de datos.
Las&nbsp;<b>relaciones</b>&nbsp;en las que se fundamenta podrían considerarse
en forma lógica como conjuntos de datos llamados&nbsp;<b>tuplas</b>.</p>

<p class=MsoNormal style='text-align:justify'>En este modelo, el lugar y la
forma en que se almacenen los datos no tienen relevancia (a diferencia de otros
modelos como el jerárquico y el de red). Esto tiene la considerable ventaja de
que es más fácil de entender y de utilizar para un usuario esporádico de la
base de datos. La información puede ser recuperada o almacenada mediante&nbsp;<b>consultas</b>&nbsp;que
ofrecen una amplia flexibilidad y poder para administrar la información.</p>

<p class=MsoNormal style='text-align:justify'>El lenguaje más habitual para
construir las consultas a bases de datos relacionales es&nbsp;<b>SQL</b>,
Structured Query Language o Lenguaje Estructurado de Consultas, un estándar
implementado por los principales motores o sistemas de gestión de bases de
datos relacionales.</p>

<p class=MsoNormal>Durante su diseño, una base de datos relacional pasa por un
proceso conocido como&nbsp;<b>normalización</b>&nbsp;de una base de datos.</p>

<ul style='margin-top:0cm' type=disc>
 <li class=MsoNormal style='text-align:justify'><b>Modelo No Relacional o
     NoSQL. </b>Las bases de datos NoSQL están orientadas a los documentos y
     permiten almacenar y recuperar datos en formatos que no sean tablas.</li>
</ul>

<p class=MsoNormal style='text-align:justify'>No utilizan SQL como lenguaje de
consultas ni estructuras fijas como tablas para el almacenamiento de los datos.</p>

<p class=MsoNormal align=center style='text-align:center'><span lang=ES-TRAD><a
href="./imagenes/sql_nosql.png"
target="_blank"><b><span lang=ES style='color:windowtext;text-decoration:none'><img
border=0 width=567 height=342 id="Imagen 2"
src="./imagenes/image005.png"></span></b></a></span></p>

<p class=MsoNormal align=center style='text-align:center'><i>Modelo SQL vs
NoSQL</i></p>

<p class=MsoNormal align=center style='text-align:center;border:none;
padding:0cm'><b><span style='font-size:16.0pt;line-height:107%;color:#0070C0'>Diseño 
de bases de datos</span></b></p>
</div>

<p class=MsoNormal>Es un proceso fundamental para modelar el conjunto de datos
y las relaciones entre ellos. Una base de datos bien diseñada influye de forma
directa en la eficiencia obtenida a la hora de almacenar, recuperar y analizar
nuestros datos.</p>

<p class=MsoNormal>Para extraer estos datos, se debe realizar un análisis en
profundidad del problema, para averiguar qué datos son esenciales para la base
de datos y descartar los que no sean necesarios. Una vez extraídos los datos
esenciales comenzamos a construir los modelos adecuados que expresen con total
exactitud todos los datos que el problema requiere almacenar.</p>

<p class=MsoNormal><b>Ventajas</b></p>

<ul style='margin-top:0cm' type=disc>
 <li class=MsoNormal>Ahorro de espacio ya que se evitan datos duplicados.</li>
 <li class=MsoNormal>Precisión e integridad de los datos. No se pierde
     información.</li>
 <li class=MsoNormal>Agilidad en el acceso y procesamiento de los datos.</li>
</ul>

<p class=MsoNormal><b>Fases</b></p>

<ol style='margin-top:0cm' start=1 type=1>
 <li class=MsoNormal><b>Análisis de requerimientos</b>. Se recopila información
     como los datos que se quieren almacenar, los usuarios de la base datos y
     sus necesidades a la hora de acceder a dichos datos. El posterior análisis
     de esta información determinará el diseño de la base de datos.</li>
</ol>

<p class=MsoNormal>Para ello, típicamente los informáticos se reúnen con los
futuros usuarios del sistema para recopilar la información que necesitan para
saber que desean dichos usuarios. Normalmente se hace una reunión inicial a y
partir de ella se elabora una batería de preguntas para entrevistar a los
usuarios finales en una segunda reunión y obtener de ella una información
detallada de lo que se espera de la base de datos. De estas entrevistas, se
extrae el documento más importante del análisis, el documento de&nbsp;<b>Especificación
de Requisitos Software</b>&nbsp;o&nbsp;<b>E.R.S.</b></p>

<ol style='margin-top:0cm' start=2 type=1>
 <li class=MsoNormal><b>Diseño conceptual</b>. Se crea una descripción a alto
     nivel de la base de datos, siguiendo el modelo entidad-relación y
     definiendo las entidades, atributos y relaciones.</li>
</ol>

<p class=MsoNormal>El objetivo de esta fase del diseño consiste es representar
la información obtenida del usuario final y concretada en el E.R.S. mediante
estándares para que el resto de la comunidad informática pueda entender y
comprender el modelo realizado.</p>

<p class=MsoNormal>El modelo que se utiliza en esta primera fase del diseño
tiene un gran poder expresivo para poder comunicarse con el&nbsp;<b>usuario que
no es experto</b>&nbsp;en bases de datos y se denomina&nbsp;<b>Modelo
Conceptual</b>.</p>

<p class=MsoNormal>Ejemplo:</p>

<p class=MsoNormal align=center style='text-align:center'><span lang=ES-TRAD><a
href="./imagenes/fase2.png"
target="_blank"><span lang=ES style='color:windowtext;text-decoration:none'><img
border=0 width=261 height=141 id="Imagen 20"
src="./imagenes/image006.png"></span></a></span></p>

<ol style='margin-top:0cm' start=3 type=1>
 <li class=MsoNormal><b>Diseño lógico</b>. Traduce el diseño conceptual a la
     que va a ser la estructura de la base de datos, definiendo las tablas, con
     sus filas y columnas, y las relaciones en nuestro caso, ya que vamos a
     seguir un modelo relacional.</li>
</ol>

<p class=MsoNormal>Este modelo es&nbsp;<b>más técnico</b>&nbsp;que el anterior
porque está orientado al personal informático y generalmente tiene traducción
directa al modelo físico que entiende el SGBD.</p>

<p class=MsoNormal>Ejemplo:</p>

<p class=MsoNormal align=center style='text-align:center'><span lang=ES-TRAD><a
href="./imagenes/fase3.png"
target="_blank"><span lang=ES style='color:windowtext;text-decoration:none'><img
border=0 width=451 height=121 id="Imagen 19"
src="./imagenes/image007.png"></span></a></span></p>

<ol style='margin-top:0cm' start=4 type=1>
 <li class=MsoNormal><b>Diseño físico</b>. A partir del modelo lógico define la
     estructura de la base de datos de forma física, es decir, en forma de
     código. En nuestro caso, SQL.</li>
</ol>

<p class=MsoNormal>Ejemplo:</p>

<p class=MsoNormal><i>CREATE TABLE PERSONAS (DNI Text PRIMARY KEY, Nombre Text)</i></p>

<p class=MsoNormal><i>CREATE TABLE TELEFONOS (Identificador Integer PRIMARY KEY
Autoincrement, DNIPersona Text, Telefono Text, FOREIGN KEY (DNIPersona)
References PERSONAS(DNI))</i></p>

<ol style='margin-top:0cm' start=5 type=1>
 <li class=MsoNormal><b>Implementación</b>. Creación y compilación para generar
     los ficheros y aplicaciones que permiten el uso práctico de la base de
     datos.</li>
</ol>

<p class=MsoNormal>Ejemplo:&nbsp;<i>agenda.db</i></p>

<p class=MsoNormal><b>Conceptos importantes</b></p>

<ul style='margin-top:0cm' type=disc>
 <li class=MsoNormal><b>Integridad</b>. Es el término usado para referirse a la
     exactitud y fiabilidad de los datos. Mantener la integridad de los datos
     significa asegurarse de que los datos permanezcan intactos y sin cambios a
     lo largo de todo su ciclo de vida.</li>
</ul>

<p class=MsoNormal align=center style='text-align:center;border:none;
padding:0cm'><b><span style='font-size:16.0pt;line-height:107%;color:#0070C0'>Tipos de integridad de datos</span></b></p>

</div>

<ul style='margin-top:0cm' type=disc>
 <ul style='margin-top:0cm' type=circle>
  <li class=MsoNormal><b>Integridad de la entidad</b>: define que cada fila sea
      única dentro de su tabla. No hay dos filas iguales. Para lograr esto, se
      puede definir una clave principal. El campo de clave principal contiene
      un identificador único, o sea, no hay dos filas que puedan contener el
      mismo identificador único.</li>
 </ul>
</ul>

<p class=MsoNormal align=center style='text-align:center'><span lang=ES-TRAD><a

<p class=MsoNormal align=center style='text-align:center'><i>Integridad de la
Entidad</i></p>

<ul style='margin-top:0cm' type=disc>
 <ul style='margin-top:0cm' type=circle>
  <li class=MsoNormal><b>Integridad referencial</b>: se refiere a las
      relaciones. Cuando dos o más tablas tienen una relación, debemos
      asegurarnos de que el valor de la clave externa coincida con el valor de
      la clave primaria en todo momento.</li>
 </ul>
</ul>

<p class=MsoNormal align=center style='text-align:center'><span lang=ES-TRAD><a
href="./imagenes/integr_ref.png"
target="_blank"><span lang=ES style='color:windowtext;text-decoration:none'><img

<p class=MsoNormal align=center style='text-align:center'><i>Integridad
Referencial</i></p>

<ul style='margin-top:0cm' type=disc>
 <ul style='margin-top:0cm' type=circle>
  <li class=MsoNormal><b>Integridad de dominio</b>: se refiere a la validez de
      las entradas para una columna determinada. Seleccionar el tipo de datos
      apropiado para una columna es el primer paso para mantener la integridad
      del dominio.</li>
 </ul>
</ul>

<p class=MsoNormal>Pero también se puede tener en cuenta la configuración de
restricciones, las reglas apropiadas para definir el formato de datos o
restringir el rango de valores posibles de entrada.</p>

      usuario aplicar reglas a la base de datos que no están cubiertas por

<p class=MsoNormal style='text-align:justify'>Puede decirse que cada tipo de

<p class=MsoNormal align=center style='text-align:center'><span lang=ES-TRAD><a

<ul style='margin-top:0cm' type=disc>
 <li class=MsoNormal><b>Consistencia</b>. Es la seguridad de que la información
     está completa (los datos se mantienen idénticos) durante cualquier
     operación, como transferencia, almacenamiento y recuperación, es decir, la
     información es consistente y correcta en cualquier momento.</li>
</ul>

<p class=MsoNormal>Como ejemplo supongamos que se tiene que grabar un pedido y
para ello se existen dos tablas:&nbsp;<i>Cabecera de Pedidos</i>&nbsp;y&nbsp;<i>Detalle
de Pedidos</i>. Se graba la cabecera y antes de que se grabe todo el detalle
ocurre un corte en la energía eléctrica, se daña la red o algún otro problema
que impide grabar todo el detalle de dicho pedido, por lo que este se queda
incompleto.</p>

<p class=MsoNormal>En este punto, es necesario conocer el concepto de&nbsp;<b>transacción</b>.
Las transacciones son unidades o secuencias de trabajo realizadas de forma
ordenada y separada en una base de datos, es decir, está compuesta por varios
procesos que se han de aplicar uno después del otro. En el ejemplo anterior,
una transacción empieza cuando se inicia la grabación del pedido y finaliza
tras insertar todo el detalle.</p>

<p class=MsoNormal>El error de consistencia de los datos puede ser detectado y
corregido enseguida, pero a veces pueden pasar días, meses o incluso años antes
de ser detectado. Este problema causaba serios perjuicios a las empresas,
motivo por el cual muchas personas se dedicaron a investigar como
solucionarlos. A finales de los años 70 Jim Gray definió las propiedades que
debía tener una transacción confiable y en 1983 Andreas Reuter y Theo Härder
inventaron la palabra&nbsp;<b>ACID</b>&nbsp;para describir a ese tipo de
transacción.</p>

<p class=MsoNormal align=center style='text-align:center'><span lang=ES-TRAD><a
href="imagenes/acid.png"
target="_blank"><span lang=ES style='color:windowtext;text-decoration:none'><img
border=0 width=567 height=313 id="Imagen 13"
src="imagenes/image013.png"></span></a></span></p>

<p class=MsoNormal style='text-align:justify'><b>ACID</b>&nbsp;es el acrónimo
de&nbsp;<b>Atomicity</b>,&nbsp;<b>Consistency</b>,&nbsp;<b>Isolation</b>,&nbsp;<b>Durability</b>&nbsp;(en
castellano: Atomicidad, Consistencia, Aislamiento, Durabilidad)</p>

<p class=MsoNormal style='text-align:justify'><b>Atomicidad</b>. Las
modificaciones a la Base de Datos (agregar, borrar o modificar) deben seguir la
regla de todo o nada. Si cualquier parte de la transacción falló, por más
pequeña que sea dicha parte, toda la transacción falló.</p>

<p class=MsoNormal style='text-align:justify'><b>Consistencia</b>. Solamente
datos válidos pueden ser escritos en la Base de Datos. Dentro de una
transacción se pueden tener temporalmente datos inválidos pero cuando la
transacción termina, sea de la forma que sea, la Base de Datos debe tener
solamente datos válidos.</p>

<p class=MsoNormal style='text-align:justify'><b>Aislamiento</b>. Si varias
transacciones se están ejecutando al mismo tiempo, ninguna de ellas interfiere
con las otras. Eso implica que la transacción A no conoce ni puede cambiar lo
que la transacción B está haciendo. Cada transacción está aislada de todas las
demás transacciones, o sea que cada transacción es totalmente independiente de
todas las demás transacciones.</p>

<p class=MsoNormal style='text-align:justify'><b>Durabilidad</b>. Cuando los
cambios a los datos que realizó la transacción son grabados, dichos cambios
permanecerán aun cuando se corte la energía eléctrica, se interrumpa la
conexión a la red, o cualquier otro problema físico.</p>

<p class=MsoNormal>Si una Base de Datos no cumple con alguno de esos 4
criterios entonces no puede ser considerada confiable.</p>

<ul style='margin-top:0cm' type=disc>
 <li class=MsoNormal style='text-align:justify'><b>Redundancia</b>. La
     redundancia de datos es la copia (duplicidad) de los datos, que se produce
     cuando algunas piezas o porciones de datos se almacenan dos o más veces.
     La redundancia de datos también tiene lugar si ciertos datos se pueden
     derivar directamente de otros.</li>
</ul>

<p class=MsoNormal align=center style='text-align:center'><span lang=ES-TRAD><a

target="_blank"><span lang=ES style='color:windowtext;text-decoration:none'><img

src="imagenes/image014.png"></span></a></span></p>

<p class=MsoNormal align=center style='text-align:center'><i>Redundancia</i></p>

<p class=MsoNormal style='text-align:justify'>Puede que la redundancia no pueda
causar problemas tan graves como los conceptos anteriores, pero está haciendo
que la base de datos sea más grande y ocupe más espacio, y además, agrega un
problema, si se quiere actualizar esta información no se puede actualizar solo
en un lugar para evitar que la información sea inconsistente.</p>

<p>El lenguaje de programación SQL se organiza en varios componentes principales, que incluyen:</p>

<p>
<b style="color:blue;">1. DDL (Data Definition Language) - Lenguaje de Definición de Datos</b></p>
<ul>
    <li><strong>Qué es</strong>: DDL se utiliza para definir la estructura de una base de datos.</li>
    <li><strong>Función</strong>: Te permite crear, modificar y eliminar tablas y otros objetos en la base de datos.</li>
    <li><strong>Ejemplos</strong>:
        <ul>
            <li><code>CREATE TABLE</code>: Crea una nueva tabla.</li>
            <li><code>ALTER TABLE</code>: Modifica una tabla existente.</li>
            <li><code>DROP TABLE</code>: Elimina una tabla.</li>
        </ul>
    </li>
    <li><strong>Por qué es importante</strong>: Sin DDL, no podrías establecer la estructura donde almacenarás tus datos.</li>
</ul>

<p><b style="color:blue;">2. DML (Data Manipulation Language) - Lenguaje de Manipulación de Datos</b></p>
<ul>
    <li><strong>Qué es</strong>: DML se utiliza para manipular los datos dentro de las tablas.</li>
    <li><strong>Función</strong>: Te permite insertar, actualizar,  y eliminar datos.</li>
    <li><strong>Ejemplos</strong>:
        <ul>
            <li><code>INSERT INTO</code>: Agrega nuevos registros.</li>
            <li><code>UPDATE</code>: Modifica registros existentes.</li>
            <li><code>DELETE</code>: Elimina registros.</li>
        </ul>
    </li>
    <li><strong>Por qué es importante</strong>: DML es esencial para interactuar con los datos almacenados en la base de datos.</li>
</ul>

<p><b style="color:blue;">3. DCL (Data Control Language) - Lenguaje de Control de Datos</b></p>
<ul>
    <li><strong>Qué es</strong>: DCL se utiliza para controlar el acceso a los datos en la base de datos.</li>
    <li><strong>Función</strong>: Te permite otorgar y revocar permisos a los usuarios.</li>
    <li><strong>Ejemplos</strong>:
        <ul>
            <li><code>GRANT</code>: Otorga permisos a un usuario.</li>
            <li><code>REVOKE</code>: Revoca permisos de un usuario.</li>
        </ul>
    </li>
    <li><strong>Por qué es importante</strong>: DCL asegura que solo las personas autorizadas puedan acceder o modificar los datos, lo que es crucial para la seguridad de la base de datos.</li>
</ul>

<p><b style="color:blue;">4. TCL (Transaction Control Language) - Lenguaje de Control de Transacciones</b></p>
<ul>
    <li><strong>Qué es</strong>: TCL se utiliza para gestionar transacciones en la base de datos.</li>
    <li><strong>Función</strong>: Te permite agrupar varias operaciones en una sola transacción, asegurando que se completen todas o ninguna.</li>
    <li><strong>Ejemplos</strong>:
        <ul>
            <li><code>COMMIT</code>: Guarda los cambios realizados en la transacción.</li>
            <li><code>ROLLBACK</code>: Revierte los cambios si hay un error.</li>
        </ul>
    </li>
    <li><strong>Por qué es importante</strong>: TCL garantiza la integridad de los datos al asegurar que las transacciones se completen correctamente.</li>
</ul>

<p><b style="color:blue;">5. DQL (Data Query Language) - Lenguaje de Consulta de Datos</b></p>
<ul>
    <li><strong>Qué es</strong>: DQL se utiliza para consultar datos en la base de datos.</li>
    <li><strong>Función</strong>: Te permite seleccionar y recuperar datos específicos.</li>
    <li><strong>Ejemplo</strong>: 
        <ul>
            <li><code>SELECT</code>: Recupera datos de una o más tablas.</li>
        </ul>
    </li>
    <li><strong>Por qué es importante</strong>: DQL es fundamental para obtener la información que necesitas de la base de datos.</li>
</ul>

<p>Resumen</p>
<ul>
    <li>Estos lenguajes son importantes porque cada uno tiene un propósito específico en la gestión de bases de datos.</li>
    <li>Te permiten crear la estructura (DDL), manipular los datos (DML), controlar el acceso (DCL), gestionar las transacciones (TCL) y consultar la información (DQL).</li>
    <li>Sin estos lenguajes, sería muy difícil trabajar de manera eficiente y segura con bases de datos.</li>
</ul>



<p>¿Entonces <code>SQL</code> esta compuestos por lenguajes?</p>

<p>Sí, exactamente. SQL (Structured Query Language) está compuesto por varios lenguajes o sub-lenguajes que se utilizan para diferentes propósitos al interactuar con bases de datos. Cada uno de estos lenguajes se especializa en una función particular, lo que permite a los usuarios realizar diversas operaciones sobre los datos y la estructura de la base de datos de manera eficiente.</p>

<h4>¿Por qué SQL está compuesto por diferentes lenguajes?</h4>
<ul>
    <li><strong>Especialización</strong>: Cada sub-lenguaje está diseñado para un conjunto específico de operaciones, lo que facilita el trabajo con bases de datos complejas.</li>
    <li><strong>Organización</strong>: Al dividir las funcionalidades en diferentes lenguajes, se mantiene un enfoque más organizado y claro para los desarrolladores y administradores de bases de datos.</li>
    <li><strong>Eficiencia</strong>: Permite a los usuarios realizar operaciones específicas sin tener que recordar todos los comandos posibles de SQL, haciendo que el trabajo sea más eficiente.</li>
</ul>

<h4>Resumen</h4>



<h5>Para evitar que podamos duplicar la creación de dos tablas con el mismo nombre se utiliza la cláusula <code>IF NOT EXISTS</code></h5>
<p>A continuación se presenta la sintaxis básica:</p>

<pre><code>CREATE TABLE IF NOT EXISTS nombre_de_la_tabla (
    campo1 TIPO_DE_DATO CONSTRAINTS,
    campo2 TIPO_DE_DATO CONSTRAINTS,
    ...
    campoN TIPO_DE_DATO CONSTRAINTS
);</code></pre>


````python
%%sql 
-- Create Table 
CREATE TABLE IF NOT EXISTS Persona(
    Dni VARCHAR(9) PRIMARY KEY,
    Nombre VARCHAR(30) NOT NULL,
    Edad INTEGER
);
````

````python
%sql SELECT * FROM Persona;
````

````python
%%sql
INSERT INTO Persona (Dni, Nombre, Edad)
VALUES 
(123457,"PEPE",23);
````


<p><b>Explicación de la Sintaxis:</b></p>
<P>La tabla se creara <b>si no está ya creada</b></P>
<ul>
    <li><b>nombre_de_la_tabla:</b> El nombre de la tabla que se va a crear.</li>
    <li><b>campo1, campo2, ..., campoN:</b> Los nombres de los campos o columnas que se van a definir dentro de la tabla.</li>
    <li><b>TIPO_DE_DATO:</b> El tipo de dato que almacenará la columna, como <code>INT</code>, <code>VARCHAR</code>, <code>BOOLEAN</code>, etc.</li>
    <li><b>CONSTRAINTS:</b> Restricciones opcionales, como <code>PRIMARY KEY</code>, <code>NOT NULL</code>, <code>UNIQUE</code>, etc., que definen las reglas para cada campo.</li>
</ul>

   Las **restricciones** (o **constraints**) en SQL son reglas opcionales que puedes aplicar a los campos de una tabla para asegurar la integridad y la validez de los datos. Aquí están algunas de las más comunes:

   1. **`PRIMARY KEY`**: Establece una columna o un conjunto de columnas como clave primaria, lo que garantiza que cada valor sea único y no nulo. Esto ayuda a identificar de manera única cada registro en la tabla.
      ```sql
      id INT PRIMARY KEY
      ```

   2. **`NOT NULL`**: Obliga a que una columna no pueda tener valores nulos, asegurando que siempre se proporcione un valor.
      ```sql
      nombre VARCHAR(50) NOT NULL
      ```


````python
%%sql
INSERT INTO Persona (Dni, Nombre, Edad)
VALUES 
(123457,"PILAR",23),
(456789, NULL, 28);
````

La sentencia anterior nos da un <b style="color:red;">error</b>, ya que definimos la tabla para que no admitiera valores `NULL` Para poder ver la estructura interna de la tabla, se utiliza el comando `PRAGMA table_info` en SQLite permite ver la estructura de la tabla, mostrando todos los detalles de sus columnas, como el nombre, tipo de datos, restricciones, si permite valores nulos, y si es clave primaria.

````python
%%sql
PRAGMA table_info(Persona);
````

   3. **`UNIQUE`**: Restringe una columna a que solo contenga valores únicos, previniendo duplicados, aunque permite valores nulos (con ciertas variaciones en algunos motores de base de datos).
      ```sql
      correo VARCHAR(100) UNIQUE
      ```


````python
%%sql 
-- Create Table 
CREATE TABLE IF NOT EXISTS ClientePrueba(
    Dni VARCHAR(9) PRIMARY KEY,
    Nombre VARCHAR(30) NOT NULL,
    Ciudad VARCHAR(20) UNIQUE,
    Edad INTEGER
);

INSERT INTO ClientePrueba (Dni, Nombre, Ciudad, Edad)
VALUES 
(123457,"PEPE", "Toledo", 23);
````

````python
%%sql
SELECT * FROM ClientePrueba;
````

Y si volvemos a introducir de nuevo el valor `Toledo`, nos devuelve un <b style="color:red;">error</b>, ya que definimos el campo como **`UNIQUE`**.

````python
%%sql
INSERT INTO ClientePrueba (Dni, Nombre, Ciudad, Edad)
VALUES 
(123887,"LUIS", "Toledo", 23);
````


   5. **`FOREIGN KEY`**: Establece una relación entre una columna en una tabla y una columna en otra tabla, asegurando que los valores coincidan en ambas.
      ```sql
      cliente_id INT, FOREIGN KEY (cliente_id) REFERENCES clientes(id)
      ```

   Estas restricciones aseguran que los datos en una base de datos mantengan un alto nivel de precisión y consistencia, facilitando también la integridad referencial y reglas específicas para los datos.
